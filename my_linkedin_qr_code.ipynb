```python
import qrcode 
import matplotlib.pyplot as plt 
  
```

```python
# String which represent the QR code 

pragati_linkedin= "https://www.linkedin.com/in/pragati-prajapati-21577516b/"
  
filename="my_qr_code.png"

image=qrcode.make(pragati_linkedin)

image.save(filename)
plt.imshow(image,cmap='gray')
```
